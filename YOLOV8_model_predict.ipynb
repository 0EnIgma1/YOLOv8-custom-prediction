## Import YOLO from ultralytics

```python
import cv2
import os
from ultralytics import YOLO
from ultralytics.yolo.utils.plotting import Annotator
import numpy as np
from google.colab.patches import cv2_imshow
from ultralytics.yolo.v8.detect.predict import DetectionPredictor
from google.colab.patches import cv2_imshow
```

## Using the pre Trained weights of the yolov8 model

```python
model = YOLO('yolov8x.pt')
```

## Trained using COCO Dataset, passing only specific classes index to detect and ignore other classes

```python
#Person and Vehicle Detetction
source = '/content/test_image.jpg'
result = model(source, classes = [0,1,2,3,5,7,15,16],conf = 0.4)
image = result[0].plot()
cv2_imshow(image)
```

```python
#Pet animals Detetction
source = '/content/test_image2.jpg'
result = model(source, classes = [0,1,2,3,5,7,15,16],conf = 0.4)
image = result[0].plot()
cv2_imshow(image)
```
